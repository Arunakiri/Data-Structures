## Minimum Coin Change

#### Dynamic Programming

```python
import math

def coin_change(target, coins):
    ways = [0]+[math.inf]*target
    
    for coin in coins:
        for i in range(coin, target+1):
            ways[i] = min(ways[i-coin]+1, ways[i])
    return ways[-1]
```

```python
target = 23
coins = [1,2,5,10,20]

coin_change(target, coins)
```

```python
def rec_coin(target, coins):
    
    min_coins = target
    
    if target in coins:
        return 1
    else:
        for i in [c for c in coins if c<=target]:
            num_coins = 1+rec_coin(target-i,coins)
            min_coins = min(num_coins, min_coins)

    return min_coins
```

```python
target = 73
coins = [1,2,5,10,20,50]

rec_coin(target, coins)
```

### Recursion Dynamic

```python
def rec_coin_dynm(target, coins, known_results):
    min_coins = target
    
    if target in coins:
        known_results[target] = 1
        return 1
    
    elif known_results[target] > 0:
        return known_results[target]
    
    else:
        for i in [c for c in coins if c <= target]:
            num_coins = 1 + rec_coin_dynm(target-i, coins, known_results)
            min_coins = min(num_coins, min_coins)
            known_results[target] = min_coins
    
    return min_coins
```

```python
target = 73
coins = [1,2,5,10,20,50]
known_results = [0]*(target+1)

rec_coin_dynm(target, coins, known_results)
```
